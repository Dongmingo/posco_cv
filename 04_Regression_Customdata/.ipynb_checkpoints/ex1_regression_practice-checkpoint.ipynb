### Custom Dataset
다음의 세가지 attribute는 반드시 정의
- __init__(self, *args, **kargs)
- __len__(self)
- __getitem__(self, idx)

```python
class FaceDataset(Dataset):
    def __init__(self, root_dir, transform):
        super().__init__()
        # implement this.
        # 필요한 정보들 생성.
        # Hint: os.listdir(directory) -> List[filename]
        self.root_dir = root_dir
        self.files = os.listdir(root_dir)
        self.transform = transform
   
    def __len__(self):
        # implement this.
         #len(FaceDataset)
        # 데이터 셋의 크기가 얼마인지? (또는, image가 몇장 있는지)
        return len(self.files)
        
    def __getitem__(self, idx):
        # implement this
        # 인덱스 (idx)를 받아서, 해당 idx 의 data 와 label 을 반환.
        # Hint: Image.open(filename).convert('RGB')
        # Hint: '1' -> 1. 로 바꾸기 위해선 float() 을 사용.
        # Hint: filename = {age}_{gender}_{race}_{time}.jpg
        file_path = os.path.join(self.root_dir, self.files[idx])
        filename = self.files[idx]
        img = Image.open(file_path).convert('RGB')
        img = self.transform(img)
        sp = self.files[idx].split('_')
        age = float(sp[0])
        gender = float(sp[1])
        race = float(sp[2])
        
        return {
            "image": img, # [3, 224, 224]
            "age": age,
            "gender": gender,
            "race": race,
            "filename": filename
        }
```

```python
# implement dataset and dataloader, train, test split maybe val set
dataset = FaceDataset("./crop_part1", transform)
```

```python
train_length = int(len(dataset) * 0.9)
test_length = len(dataset) - train_length
train_set, test_set = torch.utils.data.random_split(dataset, [train_length, test_length])
print(len(train_set), len(test_set))
train_set[0]
train_set.__getitem__(0)
```

### Custom collate function

```python
def my_collate_fn(list_data):
#     [dataset.__getitem__(0), ..., dataset.__getitem__(7)]

#     print("My collate function is working!!!!")
    data_dicts = {
        "image": [], # torch tensor [8, 3, 224, 224]
        "age": [], # torch tensor [8]
        "gender": [], # torch tensor [8]
        "race": [], # torch tensor [8]
        "filename": [] # torch tensor [8]
    }
    
    batch_size = len(list_data)
    for i in range(batch_size):
        data = list_data[i] # {"image": torch.Tensor, "age": float, "filename": str}
        data_dicts["image"].append(data["image"].unsqueeze(0)) # [1, 3, 224, 224]
        data_dicts["age"].append(data["age"])
        data_dicts["gender"].append(data["gender"])
        data_dicts["race"].append(data["race"])
        data_dicts["filename"].append(data["filename"])
    data_dicts["image"] = torch.cat(data_dicts["image"], dim=0)
    data_dicts["age"] = torch.tensor(data_dicts["age"])
    data_dicts["gender"] = torch.tensor(data_dicts["gender"])
    data_dicts["race"] = torch.tensor(data_dicts["race"])
    return data_dicts
```

```python
batch_size = 8
trainloader = DataLoader(
    train_set, batch_size = batch_size, shuffle = True, collate_fn=my_collate_fn
)
testloader = DataLoader(
    test_set, batch_size = batch_size, shuffle = False, collate_fn=my_collate_fn
)
```

```python
# display some images

import matplotlib.pyplot as plt
import numpy as np

def imshow(img):
    img = img / 2 + 0.5  # Unnormalize
    npimg = img.numpy()
    plt.figure(figsize=(20,12))
    plt.imshow(np.transpose(npimg, (1, 2, 0)))
    plt.show()
    
    
# get some random training images
dataiter = iter(trainloader)
data_dicts = dataiter.next()

# show images
imshow(torchvision.utils.make_grid(data_dicts["image"]))

# print labels
Gender_dict = {0 : 'Male', 1 : 'Female'}
Race_dict = {0 : 'White', 1 : 'Black', 2 : 'Asian', 3: 'Indian', 4 : 'Others'}
print('Age    : ', ' '.join('%6s' % str(data_dicts["age"][j].item()) for j in range(8)))
print('Gender : ', ' '.join('%6s' % Gender_dict[int(data_dicts["gender"][j].item())] for j in range(8)))
print('Race   : ',' '.join('%6s' % Race_dict[int(data_dicts["race"][j].item())] for j in range(8)))
```

## Age Regression model

```python
# ResNet18 pretrained network 받아서 씀
class Net(nn.Module):
    def __init__(self, use_pretrained=True):
        super().__init__()
        self.net = models.resnet18(pretrained = use_pretrained) #fc layer 3 는 빼고 가져옴
        in_features = self.net.fc.in_features
        self.net.fc = nn.Linear(in_features, 1) #512, 1
        
    def forward(self, x):
        return self.net(x)
```

```python
model = Net(use_pretrained=True)
summary(model, batch_size=-1, input_size=(3, 224, 224), device='cpu')
model = model.to(device)

criterion = nn.MSELoss() # nn.SmoothL1Loss(0.5)
optimizer = optim.Adam(model.parameters(), lr=0.0001)
```

```python
def train(epoch):
    model.train()
    total_loss = 0.
    for batch_idx, data_dict in enumerate(trainloader):
        data = data_dict["image"].to(device)
        target = data_dict["age"].to(dtype=torch.float32, device = device).view(-1,1) #[8] -> [8,1]
        
        optimizer.zero_grad()
        output = model(data)
        loss = criterion(output, target)
        loss.backward()
        total_loss += loss.item()
        optimizer.step()
        if batch_idx % 10 == 0:
            print('Train Epoch: {} [{}/{} ({:.0f}%)]\tLoss: {:.6f}'.format(
                epoch+1, batch_idx * len(data), len(trainloader.dataset),
                100. * batch_idx / len(trainloader), total_loss/(batch_idx+1)))

def test():
    model.eval()
    with torch.no_grad():
        test_loss = 0
        correct = np.zeros(1)
        total = np.zeros(1)
        for data_dict in testloader:
            target = data_dict["age"].to(dtype=torch.float32, device = device).view(-1,1)
            
            data, target = data_dict["image"].to(device), target.to(device)
            output = model(data)
            test_loss += criterion(output, target).item() # sum up batch loss
            
            correct += np.sum(np.array(target.to('cpu')) == np.round(np.array(output.to('cpu'))), 0)
            total += len(target)
            
        acc = correct/total
        test_loss /= len(testloader.dataset)
        print(f'\nTest set: Average loss: {test_loss:.4f}',
                f'Correct Labels for Ages : {acc[0]:.3f}')
```

```python
for epoch in range(5):
    train(epoch)
    test()
```

# Implement Regression model for Age / Gender / Race at once

```python
transform_new = transforms.Compose([
    transforms.Resize((224,224)),
    transforms.ToTensor(),
    transforms.Normalize((0.5,0.5,0.5), (0.5,0.5,0.5))
])
```

```python
class FaceDataset(Dataset):
    def __init__(self, root_dir, transform):
        super().__init__()
        self.root_dir = root_dir
        self.files_list = os.listdir(root_dir)
        self.transform = transform
   
    def __len__(self):

        return len(self.files_list)
        
    def __getitem__(self, idx):
        data_dict = {
            'image' ,
            'age' ,
            'gender' ,
            'race' 
        }
        filename = self.files_list[idx]
        img = Image.open(os.path.join(self.root_dir, filename)).convert('RGB')
        age, gender, race, timestamp = filename.split('_')
        img = self.transform(img)
        data_dict['image'] = img
        data_dict['age'] = float(age)
        data_dict['gender'] = float(gender)
        data_dict['race'] = float(race)
        data_dict['timestamp'] = float(timestamp)
        
        return data_dict
```

```python
def my_collate_fn(list_data):
    data_dicts = {
            'image' : [],
            'age' : [],
            'gender' : [],
            'race' : [],
            'timestamp' : []
        }

    for i in range(len(list_data)):
        data = list_data[i] # {"image": torch.Tensor, "age": float, "filename": str}
        data_dicts["image"].append(data["image"].unsqueeze(0)) # [1, 3, 224, 224]
        data_dicts["age"].append(data["age"])
        data_dicts["gender"].append(data["gender"])
        data_dicts["race"].append(data["race"])
    data_dicts["image"] = torch.cat(data_dicts["image"], dim=0)
    data_dicts["age"] = torch.tensor(data_dicts["age"])
    data_dicts["gender"] = torch.tensor(data_dicts["gender"])
    data_dicts["race"] = torch.tensor(data_dicts["race"])
    return data_dicts
```

```python
# ResNet18 pretrained network 받아서 씀
class Net(nn.Module):
    def __init__(self, use_pretrained=True):
        super().__init__()

        self.net = models.resnet18(pretrained = use_pretrained)
        in_features = self.net.fc.in_features
        self.net.fc = nn.Linear(in_features, 3) #512, 3
        
    def forward(self, x):
        ##implement this
        return self.net(x)
```

```python
model_new = Net(use_pretrained=True)
summary(model_new, batch_size=8, input_size=(3, 224, 224), device='cpu')
model_new = model_new.to(device)

criterion = nn.MSELoss()
optimizer = optim.Adam(model_new.parameters(), lr=0.0001)
```

```python
batch_size = 8
trainloader = DataLoader(
    train_set, batch_size = batch_size, shuffle = True, collate_fn=my_collate_fn
)
testloader = DataLoader(
    test_set, batch_size = batch_size, shuffle = False, collate_fn=my_collate_fn
)
```

```python
def train(epoch):
    model_new.train()
    total_loss = 0.
    for batch_idx, data_dict in enumerate(trainloader):
        data = data_dict["image"].to(device)
        target1 = data_dict["age"].to(device = device).to(dtype=torch.float32, device = device).view(-1,1)
        target2 = data_dict["gender"].to(device = device).to(dtype=torch.float32, device = device).view(-1,1)
        target3 = data_dict["race"].to(device = device).to(dtype=torch.float32, device = device).view(-1,1)
        target = torch.hstack([target1, target2, target3])
        
        optimizer.zero_grad()
        output = model_new(data)
        loss = criterion(output, target)
        loss.backward()
        total_loss += loss.item()
        optimizer.step()
        if batch_idx % 10 == 0:
            print('Train Epoch: {} [{}/{} ({:.0f}%)]\tLoss: {:.6f}'.format(
                epoch+1, batch_idx * len(data), len(trainloader.dataset),
                100. * batch_idx / len(trainloader), total_loss/(batch_idx+1)))

def test():
    model_new.eval()
    with torch.no_grad():
        test_loss = 0
        correct = np.zeros(3)
        total = np.zeros(3)
        for data_dict in testloader:
            target1 = data_dict["age"].to(dtype=torch.float32, device = device).view(-1,1)
            target2 = data_dict["gender"].to(dtype=torch.float32, device = device).view(-1,1)
            target3 = data_dict["race"].to(dtype=torch.float32, device = device).view(-1,1)
            target = torch.hstack([target1, target2, target3])
            
            data, target = data_dict["image"].to(device), target.to(device)
            output = model_new(data)
            test_loss += criterion(output, target).item() # sum up batch loss
            
            correct += np.sum(np.array(target.to('cpu')) == np.round(np.array(output.to('cpu'))), 0)
            total += len(target)
            
        acc = correct/total
        test_loss /= len(testloader.dataset)
        print(f'\nTest set: Average loss: {test_loss:.4f}',
                f'Correct Labels for Ages : {acc[0]:.3f}, Gender : {acc[1]:.3f}, Race : {acc[1]:.3f}')
```

```python
# Train
epochs = 10
for epoch in range(epochs):
    train(epoch)
```

```python
test()
```

# Age, Gender, Race regression results

```python
dataiter = iter(testloader)
data_dict = dataiter.next()
images = data_dicts['image']

imshow(torchvision.utils.make_grid(images))
print('Age    : ', ' '.join('%6s' % str(data_dicts["age"][j].item()) for j in range(8)))
print('Gender : ', ' '.join('%6s' % Gender_dict[int(data_dicts["gender"][j].item())] for j in range(8)))
print('Race   : ',' '.join('%6s' % Race_dict[int(data_dicts["race"][j].item())] for j in range(8)))
```

```python
images = images.to(device)
outputs = model_new(images)
outputs = np.array(outputs.detach().to('cpu'))
pred_ages = outputs[:,0]
pred_gender = ((outputs[:, 1])>0.5).astype(int)
race = outputs[:,2]
for idx in range(len(race)):
    val = race[idx]
    if val <= 0.5:
        race[idx] = 0
    elif val > 3.5:
        race[idx] = 4
    else:
        race[idx] = np.round(val)
pred_race = race.astype(int)
```
